```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
    {'type': 'insurance', 'number': '111', 'name': 'Писюн Адольфович'}
]

directories = {
    '1': ['111','2207 876234'],
    '2': ['11-2'],
    '3': ['10006'],
    '4': ['3232'],
    '5': [],
    '6': []
}

def is_document_exist (doc_number):
    return (doc_number in [v for v in directories.values() for v in v])

def is_shelf_exist (shelf):
    return shelf in [v for v in directories.keys() for v in v]

def search_document_in_directories (doc_number):
     return ' '.join([(items) for (items, v) in directories.items() for v in v if v == doc_number])

def show_person_in_documents(doc_number):
    if is_document_exist(doc_number) == True:
        print([item['name'] for item in documents if item['number'] == doc_number])        
    else:
        print('Такого документа нет')
        
def show_document_in_directories(doc_number):
    if is_document_exist(doc_number):
        print('Документ на полке: ',search_document_in_directories (doc_number))
    else:
        print('Такого документа нет')
        
def show_all_documents ():
    for document in documents:
        print(f"№:{document['number']} Документ - {document['type']}, Имя: {document['name']}. Документ на полке: {search_document_in_directories(document['number'])}")
        
def add_shelf (shelf):
    if is_shelf_exist(shelf):
        print('Такая полка уже есть!')
    else:
        directories.setdefault(shelf,[])
        print('Текущий перечень полок:', ' '.join(map(str,[v for v in directories.keys()])))

def prompt_to_delete ():
    shelf = str(input())
    if is_shelf_exist(shelf):
        del_shelf(shelf)
    else:
        print('Такой полки нет')


def del_shelf (shelf):
    if len(directories[shelf]) == 0:
        directories.pop(shelf)
        print('Полка удалена')
        print('Текущий перечень полок:', ' '.join(map(str,[v for v in directories.keys()])))  
    else:
        print('Нельзя удалить полку на которой есть документы')

def add_person ():
    person = {}
    person['type'] = str(input('Введите тип документа: '))
    person['number'] = str(input('Ведите номер документа: '))
    person['name'] = str(input('Введите Имя Фамилию лица: '))
    shelf = str(input('Введите полку для документа: '))
    if is_shelf_exist(shelf):  
        documents.append(person)
        directories_item = (directories[shelf]).append(person['number'])
        directories.setdefault(shelf,person['number'])
        print('\n Текущий список документов:')
        show_all_documents()
    else:
        print('Такой полки нет. Добавьте полку с помощью команды as')
    show_all_documents()

def del_person_document():
    doc_number = str(input('Введите документ для удаления: '))
    if is_document_exist(doc_number) == True:
        [documents.remove(item) for item in documents if item['number'] == doc_number]
        shelf = search_document_in_directories(doc_number)
        directories[shelf].remove(doc_number)
        directories.setdefault(shelf,directories[shelf])
        print('Инфоормация по документу удалена')
        show_all_documents()
    else:
        print('Такого документа нет')
        show_all_documents ()

        
def to_move_the_shelf():
    doc_number = str(input('Введите документ для перемещение: '))
    if is_document_exist(doc_number) == True:
        the_move_shelf = str(input('Введите полку куда переместить: '))
        if is_shelf_exist (the_move_shelf) == True:
            shelf = search_document_in_directories(doc_number)
            directories[shelf].remove(doc_number)
            directories.setdefault(shelf,directories[shelf])
            directories_item = (directories[the_move_shelf]).append(doc_number)
            directories.setdefault(the_move_shelf,doc_number)
            print('Перечень документов:')
            show_all_documents ()
        else:
            print('Такой полки не существует')
            print('Текущий перечень полок:', ' '.join(map(str,[v for v in directories.keys()])))
    else:
        print('Документа нет в базе')
        print('Перечень документов:')
        show_all_documents ()
```

```python
def main():
    while True:
        user_input = input('Введите команду:')
        if user_input == 'p':
            print('\n Поиск имя по номеру документа')
            doc_number = input('Введите номер документа')
            show_person_in_documents(doc_number)
        elif user_input == 's':
            print('\n Поиск полки по документу')
            doc_number = input('Введите номер документа: ')
            show_document_in_directories(doc_number)
        elif user_input == 'l':
            show_all_documents ()
        elif user_input == 'ads':
            print('\n Добавление новой полки')
            shelf = input('Введите номер полки: ')
            add_shelf (shelf)
        elif user_input == 'ds':
            print('\n Удаление пустой полки')
            shelf = input('Введите номер полки: ')
            del_shelf (shelf)
        elif user_input == 'ad':
            print('\n Добавление информации!')
            add_person ()
        elif user_input == 'd':
            print('\n Удалить информацию по номеру документа!')
            del_person_document()
        elif user_input == 'm':
            print('\n Переместить информацию из полки в полку!')
            to_move_the_shelf()
        elif user_input == 'q':
            print('\n Всего доброго, мразь!')
            break
        else:
            print('\n Такой команды не существует!')
```

```python
main()
```
